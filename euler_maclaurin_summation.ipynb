```python
import numpy as np
import matplotlib.pyplot as plt

def inv_square(x):
    return 1/(x*x)

xs = np.arange(10,30)
ys = inv_square(xs)

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)

ax.grid()

ax.plot(xs, ys, color='red')
ax.bar(xs[:-1], ys[:-1],
       width=1.0, color='lightblue',
       edgecolor='blue', align='edge',
       zorder=2.0)
ax.set_xticks(xs)
plt.show()
```

```python
new_ys = (ys[:-1]+ys[1:])/2.0

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)
ax.grid()

ax.plot(xs, ys, color='red')
ax.bar(xs[:-1], new_ys, width=1.0, color='lightblue', edgecolor='blue', align='edge', zorder=2.0)
ax.set_xticks(xs)
plt.show()
```

```python
diff_xs = np.arange(10, 30, 0.1)
def difference(x):
    return (-2.0 * x + 2 * np.floor(x) + 1.0)/np.power(x, 3) + inv_square(x)

fig = plt.figure(figsize=(8, 6), dpi=80)
ax = fig.add_subplot(1,1,1)

ax.plot(xs, ys, color='red')
ax.bar(xs[:-1], new_ys, width=1.0, color='lightblue', edgecolor='blue', align='edge', zorder=2.0)
ax.set_xticks(xs)
ax.plot(diff_xs, difference(diff_xs), '-o', color='green')
ax.grid()
```
